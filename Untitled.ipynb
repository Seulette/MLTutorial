```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
```

```python
data = np.load(r'C:\Users\katym\Downloads\data.npz')
images = data['images']
labels = data['labels']
```

```python
x = np.reshape(images,(730,-1))
y= labels
x_train, X_test, y_train, y_test= train_test_split(x, y, test_size = 0.2)
```

```python
pca = PCA(n_components=2)
fit_trans_X = pca.fit(x_train).transform(x_train)
```

```python
plt.figure(figsize = (7,5))
plt.scatter(fit_trans_X[:,0], fit_trans_X[:,1],c=y_train, s=20)
```

```python
from sklearn import svm
clf = svm.SVC(gamma=0.0001, C=10)
clf.fit(fit_trans_X, y_train)
```

```python
transform_testing_set=pca.transform(X_test)
y_pred = clf.predict(transform_testing_set)
print(y_pred)
from sklearn import metrics
print (metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.pipeline import Pipeline
```

```python
pca = PCA(n_components=3)
svc = svm.SVC(gamma=0.0001, C=10)
```

```python
pipeline = Pipeline([('pca', pca),('svc',svc)])
```

```python
pipeline.fit(x_train, y_train)
pipeline.score(X_test, y_test)
```

```python
from sklearn.model_selection import RandomizedSearchCV
search_range = {'pca__n_components': [5,10,15,20,30,50],
                'svc__C':[1, 30, 50, 1000, 10e3, 10e4]}
```

```python
searcher = RandomizedSearchCV(pipeline, search_range, n_iter=20)
searcher.fit(x_train, y_train)
searcher.score(X_test, y_test)
print(searcher.best_params_)
```
